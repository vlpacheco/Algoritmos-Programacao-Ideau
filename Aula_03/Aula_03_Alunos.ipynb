# **1. ESTRUTURAS DE CONTROLE E REPETICAO**

As estruturas de controle são empregadas com o objetivo de alterar o fluxo linear de uma rotina (script). Sem as estruturas de controle o programa executaria uma instrução após outra sempre no sentido **início → fim**. Com essas estruturas, torna-se possível decidir ir por diferentes caminhos, seja realizando um laço localizado, retornando ao inicio, pulando algumas instruções em função de algum evento que ocorreu, etc. Em resumo, as estruturas de controle são responsáveis por decidir quais blocos de código serão executados.

#**1.1 Condicional if, else, elif**

Estrutura condicional *if*, permite decidir se devemos ou não executar um trecho de código com base nas condições definidas. Se uma dada condição for **VERDADEIRA**, determinado trecho de código é executado. Se a condição for **FALSA**, podemos declarar *else* e indicar a execução de outro trecho de código. Tem-se, ainda, a estrutura *elif*, aplicável para quando o objetivo é testar mais de uma condição, na forma “se este caminho não for válido, este outro é”?

estrutura_if.jpg

***1.1.1 Sintaxe da estrutura if***

Exemplo 01: Testando se um determinado número é positivo através da estrutura if

```python
'''
Sintaxe: 

if (condicao):
  bloco de código
'''

num = 3

if num > 0: 
  print(num, "é um número positivo")
```

***1.1.2 Sintaxe da estrutura do if...else***

Exemplo 02: Testando se um determinado número é positivo através da estrutura if...else

```python
'''
Sintaxe if...else: 

if (condicao):
  bloco de código 1
else:
  bloco de código 2
'''

num = 10

if num >= 0:
  print(num, "é um número positivo ou é zero")
else:
  print(num, "é um número negativo")
```

***1.1.3 Sintaxe da estrutura do if...elif...else***

Exemplo 03: Teste se um determinado número é positivo, nulo ou negativo através da estrutura condicional if...elif...else.

```python
'''
Sintaxe if...elif...else: 

if (condicao):
  bloco de código 1
elif (condicao):
  bloco de código 2
else:
  bloco de código 3
'''
num = -4

if num > 0:
    print(num, "é um número positivo")
elif num == 0:
    print(num, "é zero")
else:
    print(num, "é um número negativo")
```

```python
numero = -10

if numero >= 2:
  print(f'{numero} é maior que 2')
elif numero == 1:
  print(f'{numero} é igual a 1')
else:
  print(f'{numero} é menor que 1')
```

#**1.2 Laço for**

Repete uma estrutura de código por um número definido de vezes.

laco_for.png

***Exemplos***

```python
'''
Sintaxe for: 

for i in (objeto):
  bloco de código 1
'''

#Criando uma sequência de 0 até 20 e imprimindo a mesma
x = 0
y = 20

for i in range(x, y):
  print(i)
```

```python
a = 10
b = 20

for i in range(a, b):
  print(i)
```

```python
#Criando uma lista de mercado e imprimindo a mesma
lista_do_mercado = ['atum', "pao", 'queijo']

for i in lista_do_mercado:
  print(i)

print(lista_do_mercado)
```

```python
#Listando números em um intervalo de 0 a 20, com incremento de 2

for i in range(0, 20, 2):
  print(i)
```

```python
#Pesquisando em Listas
lista_a = [5, 6, 7, 50]

#loop através da lista
for i in lista_a:
  if i == 5:
    print("Numero encontrado na lista")
```

#**1.3 Laço while**

Enquanto uma dada condição for válida, executa-se o laço contendo um bloco de código.
laco_while.png


***Exemplos***

```python
'''
Sintaxe while: 

i = i

while i (condicao):
  bloco de código 1
'''
#Print dos numeros, enquanto menores que 10

i = 1 #Contador

while i < 10:
  print(i)
  i = i + 1
```

```python
#Realizar a soma de uma sequencia de números com i = 10 até o valor n
n = 10 #Valor objetivo

soma = 0 #Operacao
i = 1 #Contador

while i <= 10:
  soma = soma + 1 
  i = i + 1

print("A soma é igual a:", soma)
```

#**1.4 Comandos break e continue combinados com estruturas de controle**

Nos exemplos anteriores, todo o código contido no loop while é executado a cada iteração. Porém, a linguagem Python fornece duas estruturas que permitem encerrar prematuramente um laço while:

**break**  → encerra de forma imediata a execução do laço. A execução do código prosse para a primeira instrução localizada após a estrutura do loop.

**continue**  → encerra a iteração do loop atual. A execução volta para o início do loop e a expressão condicional é reavaliada para determinar se o loop será executado novamente ou encerrado.


break.png.png

**Exemplo 01 - Aplicando a Estrutura break**

Nesse exemplo, o laço **while** é finalizado quando n torna-se igual a 2 e um aviso fora do laço passa a ser mostrado. Note que aqui também é utilizada a estrutura de controle if.

```python
n = 5

while n > 0:
  n = n - 1

  if n == 2:
    break
  print(n)

print('Laço finalizado')
```

**Exemplo 02 - Aplicando a Estrutura continue**

Observe que nesse exemplo, quando n = 2, o bloco de código retorna ao início. Nesse momento, a execucao nao passa pela funcao print() e, portanto, o valor 2 nao é mostrado. O laço while é interrompido apenas quando os valores tornam-se negativos.

```python
n = 5

while n > 0:
  n = n - 1

  if n == 2:
    continue
  print(n)

print('Laço finalizado')
```

**Exemplo 03 - Aplicando a Estrutura continue e pass**

Para esse exemplo, vamos criar um contador que inicia em 0 e termina em 10. Utilizaremos a funcao break parando a contagem com o contador atingir o numero 4 e faremos a mesma implementacao com a funcao pass, analisando a diferenca entre ambas.

```python
#Contador
contador = 0

while contador < 10:
  if contador == 4:
    break
  else:
    pass #ou continue
  print(contador)
  contador = contador + 1
```
