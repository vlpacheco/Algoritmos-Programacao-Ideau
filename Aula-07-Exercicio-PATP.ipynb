# 1) Aplicacao Tensao x Deformacao - Alumínio

Considere a aplicacao do ensaio de ruptura por tracao de um corpo de prova de aluminio, com diametro de 12,8 mm e comprimento inicial de 50,8 mm. Com os resultados encontrados no arquivo: "aluminio_1.csv". Para isso, faça:

a) Carregue os dados no ambiente Google Colab;

b) Converta a carga de N para KN;

c) Encontre a tensao equivalente em MPa;

d) Encontre a deformacao linear (ΔL/L) (ε) e o (ε) em %;

e) Plote o graficos: N vs deformacao (mm); KN vs deformacao (ε) em %; MPa vs deformacao (ε) em %.

```python
#Carregando as bibliotecas
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#a
#Carregando a planilha para o ambiente Colab
aluminio_df = pd.read_csv('aluminio_1.csv')
aluminio_df

#Se carregou com Unnamed
#aluminio_df = aluminio_df.drop(columns=['Unnamed: 2', 'Unnamed: 3', 'Unnamed: 4', 'Unnamed: 5', 'Unnamed: 6', 'Unnamed: 7'])

#aluminio_df
```

```python
#Filtrando colunas
#aluminio_df.loc[:, ['Carga (N)', 'Deformacao (mm)']]
aluminio_df.loc[4, ['Carga (N)', 'Deformacao (mm)']]
```

```python
#b
#Convertendo uma coluna // Operacoes com coluna
aluminio_df['Carga (KN)'] = aluminio_df['Carga (N)'] / 1000
aluminio_df

#Dropar colunas
#aluminio_df.drop(columns = 'KN')
```

```python
#c
#Calculando a area
area = (np.pi * (12.8 ** 2))/4 #mm2
area

#MPa =. N/mm2
aluminio_df['Tensao (MPa)'] = (aluminio_df['Carga (N)'] / area) #Sem arredondar
aluminio_df['Tensao (MPa)'] = aluminio_df['Tensao (MPa)'].round(2) #Arredondando as casas
#aluminio_df['Tensao (MPa)'] = round((aluminio_df['Carga (N)'] / area), 2) #Arredondar
aluminio_df
```

```python
#Rearange order
aluminio_df = aluminio_df[['Carga (N)', 'Carga (KN)', 'Tensao (MPa)','Deformacao (mm)']]
aluminio_df
```

```python
#d Deformacao Linear (ε)
#(ε) = (ΔL/L) 
#L0 = 50.8mm
Lo = 50.8
aluminio_df['Deformacao Linear (ΔL/L) (ε)'] = (aluminio_df['Deformacao (mm)'] / Lo)
aluminio_df['Deformacao Linear (ΔL/L) (ε)'] = aluminio_df['Deformacao Linear (ΔL/L) (ε)'].round(2)
aluminio_df 
```

```python
# Encontrando o (ε) (%) 
aluminio_df['Deformacao Linear (ΔL/L) (ε) (%)'] = aluminio_df['Deformacao Linear (ΔL/L) (ε)'] * 0.01
aluminio_df['Deformacao Linear (ΔL/L) (ε) (%)'] = aluminio_df['Deformacao Linear (ΔL/L) (ε) (%)'].round(6)
aluminio_df
```

```python
#e
#Plotando no Jupyter Notebook
#Adicionar o inline
%matplotlib notebook
plt.plot(aluminio_df['Deformacao (mm)'], aluminio_df['Carga (N)'])
plt.show()
```

```python
#Editando o gráfico Carga (N) vs Def (mm)
%matplotlib notebook
plt.plot(aluminio_df['Deformacao (mm)'], aluminio_df['Carga (N)'])
plt.xlabel('Deformacao (mm)')
plt.ylabel('Carga (N)')
plt.title('Carga (N) x Deformacao (mm) do Aluminio')
plt.legend(['Aluminio'], loc = 'lower right')
plt.show()
plt.savefig('Tensao-Deformacao.png', dpi = 300, bbox_inches = 'tight')
```

```python
#e
#Editando o gráfico Carga (KN) vs deformacao (ε) em %
%matplotlib notebook
plt.plot(aluminio_df['Deformacao Linear (ΔL/L) (ε) (%)'], aluminio_df['Carga (KN)'], color = 'green')
plt.xlabel('(ε) (%)')
plt.ylabel('Carga (KN)')
plt.title('Carga (KN) vs deformacao (ε) em %')
plt.legend(['Aluminio'], loc = 'lower right')
plt.show()
plt.savefig('KN-ε%.png', dpi = 300, bbox_inches = 'tight')
```

```python
#e
#Editando o gráfico Tensao (MPa) vs deformacao (ε) em %
%matplotlib notebook
plt.plot(aluminio_df['Deformacao Linear (ΔL/L) (ε) (%)'], aluminio_df['Tensao (MPa)'], color = 'magenta')
plt.xlabel('(ε) (%)')
plt.ylabel('Carga (KN)')
plt.title('Carga (KN) vs deformacao (ε) em %')
plt.legend(['Aluminio'], loc = 'center right')
plt.show()
plt.savefig('MPa-ε%.png', dpi = 300, bbox_inches = 'tight')
```
