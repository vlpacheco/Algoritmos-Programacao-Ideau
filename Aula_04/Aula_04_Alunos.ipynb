# **1. IDENTACAO**

Na linguagem escrita, a indentação é definida como um espaço em branco no início de uma linha. O espaço é usado para indicar o início de um novo **parágrafo**. De forma análoga, a indentação é utilizada por linguagens de programação para indicar o **início de um novo bloco lógico **em um programa. Os termos *indent* e *outdent* são neologismos criados para descrever os processos de inserir e remover uma indentação, respectivamente. Assim, *outdent* é o ato de remover espaços e realinhar o texto com a margem da página.

Na linguagem Python, a identação é obtida a partir da inclusão de **4 espaços (tecla spacebar x 4)**, e é de fundamental importância para a estruturação do código e para a interpretação do código pela máquina. A ordem hierárquica dos blocos de código é definida através da indentação, conforme mostra a figura abaixo. A figura ilustra o processo de indentação, que é usado para estruturar os blocos lógicos da programação em Python.

# **2. FUNCOES**

Em Python, uma função é uma sequência de comandos que executa alguma tarefa e que tem um nome. A sua principal finalidade é nos ajudar a organizar programas em pedaços que correspondam a como imaginamos uma solução do problema.

A Sintaxe da funcao é:

```python
#Sintaxe
def NOME(PARÂMETROS):
    BLOCO CÓDIGO
```

Na definição de uma função, a palavra reservada no cabeçalho é **def**, que é seguida pelo nome da função e alguns parâmetros entre parênteses. A lista de parâmetros pode ser vazia ou conter qualquer número de parâmetros separados pos vírgulas. Em qualquer caso, os parênteses são obrigatórios.

Precisamos falar um pouco mais sobre parâmetros. Na definição, a lista de parâmetros é conhecida mais especificamente como os **parâmetros formais.** Esta lista de nomes descrevem o que a função precisará receber do usuário da função. Quando você usa uma função, você fornece valores aos **parâmetros formais**.


**Eexemplo 01)** Criando funcao sem parâmetro.

```python
# Definindo uma função
def primeiraFunc():
    print('Hello World')

primeiraFunc()
```

# **3. EXPRESSOES LAMBDA**

As funções **lambda** são sempre opcionais, tudo que fazemos com ela, podemos fazer também com uma função normal. O uso de expressões **lambda** permite que você possa representar a lógica de uma função sem precisar defini-la explicitamente. Isso pode deixar o código mais facil, mais claro e mais rápido de ser lido

**Exemplo 01) ** Exemplo de aplicacao da expressao lambda e diferenciacao da definicao de funcoes.

```python
# Definindo uma função - 3 linhas de código
def potencia(num):
  """Definicao de potencia"""
  result = num ** 2
  return result

potencia(2)
```

```python
# Definindo uma função - 2 linhas de código
def potencia(num):
    return num**2

potencia(2)
```

```python
# Definindo uma EXPRESSAO LAMBDA
'''
Sintaxe

expressa = lambda parametro formal: bloco de codigo
'''

potencia = lambda num: num**2

potencia(5)
```
