## 1) Aplicacao Tensao x Deformacao - Aluminum / Steel

Considerar os ensaios de tracao para os elementos de aço (steel) e alumínio (aluminium) dispostos nos arquivos ".csv". Arquivos: "aluminum6061.csv" e "steel1045.csv". Com isso realizar:

a) O carregamento dos arquivos no Jupyter Notebook;

b) A transformacao das unidades de lb para N e para KN;

c) Calcular a tensao resultante em MPa, considerando o diametro das barras de aço e de alumínio de 0.506 in;

d) Calcular o  ε (%);

e) Plotar gráficos de tensao x deformacao (MPa vs ε (%)) separadamente;

f) Plotar gráficos de tensao x deformacao (MPa vs ε (%)) no mesmo plot.

```python
#importando as bibliotecas
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#a
steel = pd.read_csv('steel1045.csv')
steel
```

```python
aluminum = pd.read_csv('aluminum6061.csv')
aluminum
```

```python
#b
#lb to N --> *4.44
#lb to KN --> *0.0044

#steel
steel['Carga (N)'] = round(steel['Carga (lb)'] * 4.44, 2)
steel['Carga (KN)'] = round(steel['Carga (lb)'] * 0.0044, 2)
steel
```

```python
#b
#lb to N --> *4.44
#lb to KN --> *0.0044

#steel
aluminum['Carga (N)'] = round(aluminum['Carga (lb)'] * 4.44, 2)
aluminum['Carga (KN)'] = round(aluminum['Carga (lb)'] * 0.0044, 2)
aluminum
```

```python
#c 
diametro = 0.506 * 25.4 

area = ((diametro ** 2)* np.pi)/4
area
```

```python
#c
steel['Tensao (MPa)'] = round(steel['Carga (N)'] / area, 2)
steel
```

```python
#c
aluminum['Tensao (MPa)'] = round(aluminum['Carga (N)'] / area, 2)
aluminum
```

```python
#d --> ε (%)
steel['ε (%)'] = round(steel['Deformacao (ε)'] * 0.01, 4)
steel
```

```python
#Reorganizar
steel = steel[['Ponto', 'Carga (lb)', 'Carga (N)', 'Carga (KN)', 'Tensao (MPa)', 'Deformacao (ε)', 'ε (%)']]
steel
```

```python
#d --> ε (%)
aluminum['ε (%)'] = round(aluminum['Deformacao (ε)'] * 0.01, 4)
aluminum
```

```python
#Reorganizar
aluminum = aluminum[['Ponto', 'Carga (lb)', 'Carga (N)', 'Carga (KN)', 'Tensao (MPa)', 'Deformacao (ε)', 'ε (%)']]
aluminum
```

```python
#e plot MPa vs ε (%)
%matplotlib notebook
#plt.plot(steel['Tensao (MPa)'], steel['ε (%)'], color = 'magenta')
plt.plot(steel['ε (%)'], steel['Tensao (MPa)'], color = 'magenta',)
plt.xlabel('(ε) (%)')
plt.ylabel('Tensao (MPa)')
plt.title('Tensao (MPa) vs deformacao (ε) em % - Steel')
plt.legend(['Steel'], loc = 'center right')
plt.show()
#plt.savefig('steel-MPa-ε%.png', dpi = 300, bbox_inches = 'tight')
```

```python
#e plot MPa vs ε (%)
%matplotlib notebook
#plt.plot(aluminum['Tensao (MPa)'], aluminum['ε (%)'], color = 'magenta')
plt.plot(aluminum['ε (%)'], aluminum['Tensao (MPa)'], color = 'turquoise')
plt.xlabel('(ε) (%)')
plt.ylabel('Tensao (MPa)')
plt.title('Tensao (MPa) vs deformacao (ε) em % - Aluminum')
plt.legend(['Aluminio'], loc = 'center right')
plt.show()
#plt.savefig('aluminum-MPa-ε%.png', dpi = 300, bbox_inches = 'tight')
```

```python
%matplotlib notebook
#plt.plot(steel['Tensao (MPa)'], steel['ε (%)'], color = 'magenta')
plt.plot(steel['ε (%)'], steel['Tensao (MPa)'], color = 'magenta')
plt.plot(aluminum['ε (%)'], aluminum['Tensao (MPa)'], color = 'turquoise')
plt.xlabel('(ε) (%)')
plt.ylabel('Tensao (MPa)')
plt.title('Tensao (MPa) vs deformacao (ε) em % ')
plt.legend(['Steel', 'Aluminum'], loc = 'lower right')
plt.show()
plt.savefig('aluminum-steel-MPa-ε%.png', dpi = 300, bbox_inches = 'tight')
```
